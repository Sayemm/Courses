```python
def answer_one():
    import pandas as pd
    import numpy as np

    #one
    df = pd.read_excel('Energy Indicators.xls', skiprows = 17, skipfooter=38) #skip first 17 rows and last 38 rows

    df = df[['Unnamed: 1','Petajoules','Gigajoules','%']]

    df.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

    df[['Energy Supply', 'Energy Supply per Capita', '% Renewable']] = df[['Energy Supply', 'Energy Supply per Capita', '% Renewable']].replace('...',np.NaN).apply(pd.to_numeric)

    df['Energy Supply'] = df['Energy Supply']*1000000

    df['Country'] = df['Country'].replace({"Republic of Korea": "South Korea",
                                           "United States of America": "United States",
                                           "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                                           "China, Hong Kong Special Administrative Region": "Hong Kong"})

    df['Country'] = df['Country'].str.replace(" \(.*\)","")

    #two
    dp = pd.read_csv('world_bank.csv', skiprows = 4)

    dp['Country Name'] = dp['Country Name'].replace({"Korea, Rep.": "South Korea", 
                                                     "Iran, Islamic Rep.": "Iran",
                                                     "Hong Kong SAR, China": "Hong Kong"})

    dp = dp[['Country Name','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
    dp.columns = ['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']

    #three
    ds = pd.read_excel('scimagojr-3.xlsx')
    ds = ds[0:15]

    tmp = pd.merge(ds, df, how = 'inner', left_on = 'Country', right_on='Country')
    final = pd.merge(tmp, dp, how = 'inner', left_on = 'Country', right_on='Country').set_index('Country')
    
    return final

answer_one()
```

```python
def answer_two():
    import pandas as pd
    import numpy as np

    #one
    df = pd.read_excel('Energy Indicators.xls', skiprows = 17, skipfooter=38) #skip first 17 rows and last 38 rows

    df = df[['Unnamed: 1','Petajoules','Gigajoules','%']]

    df.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

    df[['Energy Supply', 'Energy Supply per Capita', '% Renewable']] = df[['Energy Supply', 'Energy Supply per Capita', '% Renewable']].replace('...',np.NaN).apply(pd.to_numeric)

    df['Energy Supply'] = df['Energy Supply']*1000000

    df['Country'] = df['Country'].replace({"Republic of Korea": "South Korea",
                                           "United States of America": "United States",
                                           "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                                           "China, Hong Kong Special Administrative Region": "Hong Kong"})

    df['Country'] = df['Country'].str.replace(" \(.*\)","")

    #two
    dp = pd.read_csv('world_bank.csv', skiprows = 4)

    dp['Country Name'] = dp['Country Name'].replace({"Korea, Rep.": "South Korea", 
                                                     "Iran, Islamic Rep.": "Iran",
                                                     "Hong Kong SAR, China": "Hong Kong"})

    dp = dp[['Country Name','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
    dp.columns = ['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']

    #three
    ds = pd.read_excel('scimagojr-3.xlsx')

    tmp = pd.merge(ds, df, how = 'inner', left_on = 'Country', right_on='Country')
    final = pd.merge(tmp, dp, how = 'inner', left_on = 'Country', right_on='Country').set_index('Country')
    
    tmp1 = pd.merge(ds, df, how = 'outer', left_on = 'Country', right_on='Country')
    final1 = pd.merge(tmp1, dp, how = 'outer', left_on = 'Country', right_on='Country').set_index('Country')
    
    return len(final1)-len(final)

answer_two()
```

```python
import pandas as pd
import numpy as np

def average(row):
    data = row[np.arange(2006, 2016).astype(str)]
    
    return pd.Series({'avgGDP' : np.mean(data)})

def answer_three():
    Top15 = answer_one()
    
    Top15 = Top15.apply(average, axis = 1)

    return Top15['avgGDP'].sort_values(ascending = False)

answer_three()
```

```python
import pandas as pd
import numpy as np

def answer_four():
    Top15 = answer_one()
    average = answer_three().sort_values(ascending = False)
    country = average.index[5]
    
    return Top15.loc[country]['2015']-Top15.loc[country]['2006']

answer_four()
```

```python
def answer_five():
    Top15 = answer_one()
    return Top15['Energy Supply per Capita'].mean()

answer_five()
```

```python
def answer_six():
    Top15 = answer_one()
    
    country = Top15['% Renewable'].idxmax()
    
    return (country, Top15.loc[country]['% Renewable'])

answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    
    new = Top15['Self-citations'] / Top15['Citations']
    country = new.idxmax()
    return (country, new.loc[country])

answer_seven()
```

```python
def answer_eight():
    Top15 = answer_one()
    
    new = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    new = new.sort_values(ascending = False)
    return new.index[2]

answer_eight()
```

```python
def answer_nine():
    Top15 = answer_one()
    
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable docs per Capita'] = Top15['Citable documents'] / Top15['PopEst']
    
    ans = Top15['Citable docs per Capita'].corr(Top15['Energy Supply per Capita'])
    return ans

answer_nine()
```

```python
def answer_ten():
    Top15 = answer_one()
    
    new = Top15['% Renewable']
    HighRenew = new
    med = new.median()
    
    for i in range(len(new)):
        if new[i]>= med: HighRenew[i] = 1
        else: HighRenew[i] = 0
    
    return  HighRenew.astype(int).sort_values(ascending=True)

answer_ten()
```

```python
def answer_eleven():
    Top15 = answer_one()
    
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15 = Top15.reset_index()
    
    Top15['Continent'] = [ContinentDict[country] for country in Top15['Country']]
    Top15 = Top15.set_index('Continent')
    
    Top15 = Top15.groupby(level=0)['PopEst'].agg({'size': np.size, 'sum': np.sum, 'mean': np.mean,'std': np.std})           
    
    return Top15

answer_eleven()
```

```python
def answer_thirteen():
    Top15 = answer_one()
    
    ans = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    ans = ans.apply('{:,}'.format)
        
    return ans

answer_thirteen()
```
