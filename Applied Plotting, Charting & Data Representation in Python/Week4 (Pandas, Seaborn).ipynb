```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
plt.style.available
plt.style.use('seaborn-colorblind')
```

```python
np.random.seed(123)

df = pd.DataFrame({'A' : np.random.randn(365).cumsum(0), 
                   'B' : np.random.randn(365).cumsum(0) + 20,
                   'C' : np.random.randn(365).cumsum(0) - 20}, 
                 index = pd.date_range('1/1/2017', periods = 365))

df.head()
```

```python
df.plot(); #semicolom surpress unwanted output
```

```python
df.plot('A', 'B')
```

```python
df.plot('A', 'B', kind = 'scatter')
```

```python
df.plot.scatter('A', 'C', c = 'B', colormap = 'viridis')
```

```python
ax = df.plot.scatter('A', 'C', c = 'B', s = df.B, colormap = 'viridis')
ax.set_aspect('equal') #aspect = equal allows the viewer to easily see that the range of series A much smaller than Series C
```

```python
df.plot.box();
```

```python
df.plot.hist();
```

```python
#kernel density estimate plot
#Which are useful for visualizing an estimate of a variable's probability density function

df.plot.kde()
```

```python
df = pd.read_csv('iris.csv')

df
```

```python
#A scatter matrix is a way of comparing each column in a DataFrame to every other column in a pairwise fashion
#The scatter matrix creates scatter plots between the different variables and histograms along the diagonals.

pd.plotting.scatter_matrix(df);
```

```python
#Parallel coordinate plots are a common way of visualizing high dimensional multivariate data.
#Each variable in the data set corresponds to an equally spaced parallel vertical line. The values of each variable are then connected by lines
#between for each individual observation.

plt.figure()
pd.plotting.parallel_coordinates(df, 'Name')
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0, 10, 1000), name = 'V1')
v2 = pd.Series(2*v1 + np.random.normal(60, 15, 1000), name = 'V2')

v2
```

```python
plt.figure()

plt.hist(v1, alpha = .7, bins = np.arange(-50, 150, 5), label = 'V1')
plt.hist(v2, alpha = .7, bins = np.arange(-50, 150, 5), label = 'V2')

plt.legend()
```

```python
plt.figure()

plt.hist([v1, v2], histtype = 'barstacked', normed = True) # Normed true = normalize the histogram to form a probability density

v3 = np.concatenate((v1, v2))

sns.kdeplot(v3)
```

```python
plt.figure()

sns.distplot(v3, hist_kws = {'color': 'teal'}, kde_kws = {'color' : 'Navy'})
```

```python
sns.jointplot(v1, v2, alpha = 0.4)
```

```python
#The jointplot creates a scatterplot along the histograms for each individual variable on each axis.

grid = sns.jointplot(v1, v2, alpha = 0.4)

grid.ax_joint.set_aspect('equal')
```

```python
#Hexbin plots show the number of observations that fall within hexagonal bins.

sns.jointplot(v1, v2, kind = 'hex')
```

```python
sns.set_style('white')

sns.jointplot(v1, v2, kind = 'kde', space = 0)

#space = 0 will plot the marginal distributions directly on the border of the scatter plot.
```
