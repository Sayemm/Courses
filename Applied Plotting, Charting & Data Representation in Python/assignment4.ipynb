```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook

plt.style.use('seaborn-colorblind')
```

```python
df = pd.read_csv("bangladesh.csv")
of = pd.read_csv("other_country.csv")

print(df.head())
print(of.head())
```

```python
plt.figure(figsize = (10, 5))

plt.plot(df.Opposition, df['Win Percentage'], label = 'Bangladesh')
plt.plot(of.Opposition, of.India, label = 'India')
plt.plot(of.Opposition, of.Pakistan, label = 'Pakistan')
plt.plot(of.Opposition, of.Srilanka, label = 'Srilanka')

plt.xlim(0, 14)

plt.xlabel('Country')
plt.ylabel('Win Percentage')
plt.title('Win% comparison in Cricket with Bangladesh and other country')

plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.legend()
plt.savefig('assignment4.png')
plt.show()
```
